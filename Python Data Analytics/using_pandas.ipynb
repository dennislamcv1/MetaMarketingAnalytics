#### Question 1

Create a DataFrame called `df` from the given CSV file `student_data.csv`


```python
# Your code here

df = pd.read_csv("student_data.csv")
```

```python
# Question 1 Grading Checks

assert isinstance(df, pd.DataFrame), "Did you create a DataFrame called df?"
```

```python
df.head()
```

#### Question 2

Get the `Math Grade` in the 27th row and assign it to a variable called `math_grade`.


```python
# Your code here

math_grade = df["Math Grade"].loc[26]
```

```python
# Question 2 Grading Checks

assert math_grade != None, 'Did you define math_grade?'
```

#### Question 3

Using the `df` DataFrame, assign the `Name` column to a `Series` called `names`.


```python
# Your code here

names = pd.Series(data=df.Name)
names
```

```python
# Question 3 Grading Checks

assert isinstance(names, pd.Series), "Did you assign the Name column in a variable called names?"
```

#### Question 4

Using the `names` Series, assign the first 3 names to a Series called `first_three_names`.


```python
# Your code here

first_three_names = names[:3]

first_three_names
```

```python
# Question 4 Grading Checks

assert isinstance(first_three_names, pd.Series), "Did you assign the first 3 names to a variable called first_three_names?"
```

#### Question 5

Using the `df` DataFrame, assign the `Age` and `Math Grade` columns to a DataFrame called `age_math`.


```python
# Your code here

age_math = pd.DataFrame(data=df, columns=["Age", "Math Grade"])

age_math
```

```python
# Question 5 Grading Checks

assert isinstance(age_math, pd.DataFrame), "Did you assign the Age and Math Grade columns to a variable called age_math?"
```

#### Question 6

Using the `.loc()` method, assign the `Age` and `Math Grade` columns for the first 30 rows of `df` to a variable called `first_thirty_loc`.


```python
# Your code here

first_thirty_loc = df.loc[:29, ['Age', 'Math Grade']]

first_thirty_loc
```

```python
# Question 6 Grading Checks

assert isinstance(first_thirty_loc, pd.DataFrame), "Did you correctly assign the first 30 rows to a variable called first_thirty_loc?"
```

#### Question 7

Using the `.iloc()` method, assign the `Age` and `Math Grade` columns for the first 30 rows of `df` to a variable called `first_thirty_iloc`.


```python
# Your code here


first_thirty_iloc = df.iloc[:30, [1, 2]]

first_thirty_iloc
```

```python
# Question 7 Grading Checks

assert isinstance(first_thirty_iloc, pd.DataFrame), "Did you correctly assign the first 30 rows to a variable called first_thirty_iloc?"
```

#### Question 8

Get the odd numbered rows from `df` and assign the result to a variable called `odd_rows`.


```python
# Your code here

odd_rows = df.iloc[1::2]

odd_rows
```

```python
# Question 8 Grading Checks

assert isinstance(odd_rows, pd.DataFrame), "Did you correctly assign odd numbered rows to a variable called odd_rows? Hint: the first row is index 0."
```

#### Question 9

Get the even numbered rows from the `Name` and `English Grade` columns and assign the result to a variable called `even_rows_english`.


```python
# Your code here

even_rows_english = df.iloc[0::2, [0, 3]]

even_rows_english
```

```python
# Question 9 Grading Checks

assert isinstance(even_rows_english, pd.DataFrame), "Did you correctly assign even numbered rows to a variable called even_rows_english? Hint: the first row is index 0."
```
